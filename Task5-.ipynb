```python
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
```

```python
url = "https://www.pararius.com/apartments/amsterdam?ac=1"
driver = webdriver.Chrome(service=(ChromeService(ChromeDriverManager().install())))
driver.get(url)
```

```python
try:
    one_trust = WebDriverWait(driver, 1000).until(
        EC.presence_of_element_located((By.ID, "onetrust-accept-btn-handler"))
    )
except:
    driver.quit()

one_trust = driver.find_element(By.ID, "onetrust-accept-btn-handler")
one_trust.click()
```

```python
item_selector = "search-list__item search-list__item--listing"
title_selector = "listing-search-item__title"
price_selector = "listing-search-item__price"
```

```python
rentals = {
    "Title" : [],
    "Price" : []
}

curr_page = 1

while True:
    url = f"https://www.pararius.com/apartments/amsterdam/page-{curr_page}"
    driver.get(url)

    titles = driver.find_elements(By.XPATH, '//h2[contains(@class, "listing-search-item__title")]')
    prices = driver.find_elements(By.XPATH, '//div[contains(@class, "listing-search-item__price")]')
   
    for i in range(len(titles)):
        rentals['Title'].append(titles[i].text)
        rentals['Price'].append(int(prices[i].text.split(" ")[0][1:].replace(",", "")))


    try:
        next = driver.find_element(By.XPATH, '//a[contains(@class, "pagination__link pagination__link--next")]') 
    except:
        driver.quit()
        break
    curr_page = next.get_attribute("data-page")
```

```python
for title, price in sorted(zip(rentals["Title"], rentals["Price"]) , key=lambda item: item[1]):
    print(f"Apartment = {title}, Price = {price}")
```
