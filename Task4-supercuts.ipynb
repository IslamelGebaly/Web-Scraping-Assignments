```python
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
url = "https://www.supercuts.com/salon-directory"
```

```python

driver.get(url)
group = driver.find_elements(By.XPATH, '//ion-item-group[contains(@role, "group")]')
country_links = []
addresses = []
```

```python
for country_link in group:
    country_links.append(country_link.find_element(By.TAG_NAME, "ion-router-link").get_attribute("href").split("/")[-1])
```

```python
for country_link in country_links:
    if(country_link == "CA"):
        url = f"https://www.supercuts.com/salon-directory/{country_link}"
        driver.get(url)
    
    grid = driver.find_element(By.XPATH, "//ion-row[contains(@class, 'ion-padding-bottom md')]")
    state_list = grid.find_elements(By.TAG_NAME, "ion-router-link")

    state_urls = []
    segments = ["A-D", "E-H", "I-L", "M-P", "Q-T", "U-W", "X-Z"]
    for state in state_list:
        link = state.get_attribute("href")
        for segement in segments:
            state_url = f"https://www.supercuts.com{link}?alpha={segement}"
            state_urls.append(state_url)

    for state_url in state_urls:
        driver.get(state_url)

        items = driver.find_elements(By.XPATH, "//ion-row[contains(@class, 'salon-detail md')]")
        for i in items:
            addresses.append(i.text.replace("\n", " "))
```

```python
url = f"https://www.supercuts.com/salon-directory/CA"
driver.get(url)
    
grid = driver.find_element(By.XPATH, "//ion-row[contains(@class, 'ion-padding-bottom md')]")
state_list = grid.find_elements(By.TAG_NAME, "ion-router-link")

state_urls = []
segments = ["A-D", "E-H", "I-L", "M-P", "Q-T", "U-W", "X-Z"]
for state in state_list:
    link = state.get_attribute("href")
    for segement in segments:
        state_url = f"https://www.supercuts.com{link}?alpha={segement}"
        state_urls.append(state_url)

for state_url in state_urls:
    driver.get(state_url)

    items = driver.find_elements(By.XPATH, "//ion-row[contains(@class, 'salon-detail md')]")
    for i in items:
        addresses.append(i.text.replace("\n", " "))
```

```python
print(*addresses, sep = "\n")
```
